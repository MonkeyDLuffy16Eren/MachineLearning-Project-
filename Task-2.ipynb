```python
import pandas as pd
```

```python
data = pd.read_csv('cogniy_data.csv')
```

```python
data.head()
```

```python
data['Cuisines'].fillna('Unknown', inplace=True)
```

```python
def recommend_restaurants_simple(user_preferences, data, top_n=5):
    recommendations = []

    for index, row in data.iterrows():
        cuisines = row['Cuisines'].split(', ')
        cuisine_match = len(set(cuisines) & set(user_preferences['Cuisines'])) / len(user_preferences['Cuisines'])
        
        price_range_diff = abs(row['Price range'] - user_preferences['Price range'])
        price_range_match = 1 / (1 + price_range_diff)
        
        overall_score = cuisine_match + price_range_match
        
        recommendations.append((overall_score, row['Restaurant Name'], row['Cuisines'], row['Price range'], row['Aggregate rating']))
    
    recommendations.sort(reverse=True, key=lambda x: x[0])
    

    top_recommendations = recommendations[:top_n]
    
    top_recommendations_df = pd.DataFrame(top_recommendations, columns=['Score', 'Restaurant Name', 'Cuisines', 'Price range', 'Aggregate rating'])
    
    return top_recommendations_df
```

```python
user_preferences = {
    'Cuisines': ['indian'],
    'Price range': 5
}

```

```python
recommended_restaurants = recommend_restaurants_simple(user_preferences, data)
print(recommended_restaurants)
```
